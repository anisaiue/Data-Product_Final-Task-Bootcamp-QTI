```python
# import library
import requests
import pandas as pd
import math, time
```

```python
# definisikan terlebih dahulu kriteria berikut
parent_api = "https://api.bukalapak.com/product-reviews?"
token = "access_token=7OR6Bawa94vQP3UXjxRhnbsFxSPgV41BiMWJpe2_fchj8w" # TO DO: isi access token
description = "&has_description=true"
rating = "&ratings[]="
product_id = "&product_id=k2a7ub"
offset = "&offset="
limit = "&limit=10"

# banyak review tiap rate
total_rates = [74, 71, 49, 18, 363]
```

```python

# lakukan crawling
# data hasil crawling akan dimasukkan ke dalam pandas DataFrame
df_review = pd.DataFrame({"id_review": [],
                          "name": [],
                          "review": [],
                          "rate": []})

i = 0
# lakukan iterasi dari bintang 1-5 serta digabungkan dengan data total bintang masing-masing
for rate, total_rate in zip(range(1, 5+1, 1), total_rates):
    print("ambil data pada bintang {} dengan banyak data {}".format(rate, total_rate))
    
    # definisikan nilai max_offset
    max_offset = math.floor(int(total_rate)/10)*10
    
    for n_offset in range(0, max_offset+1, 5):
        try:
            print("  - ambil data pada offset:", n_offset)
            
            api = parent_api+token+description+rating+str(rate)+product_id+offset+str(n_offset)+limit
            
            # ambil data dari url API
            response = requests.get(api)
            data = response.json()
            
            for d in data['data']:
                # definisikan lokasi data untuk masing-masing
                id_review = i
                name = d["sender"]["name"]
                review = d["review"]["content"]
                rate = d["review"]["rate"]
                
                # masukkan data ke dalam dataframe
                df_review = df_review.append({"id_review": int(i),
                                              "name": name,
                                              "review": review,
                                              "rate": str(rate)}, ignore_index=True)
                i+=1
            time.sleep(1)
        except:
            print("Response failed! mungkin access token bermasalah")
```
